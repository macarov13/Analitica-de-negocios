0. Cargar las librerias de referencia para el modelo

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
```

1. Se procede a la carga de datos del archivo

```python
nxl='/content/drive/MyDrive/Colab Notebooks/2. BD1_Travel_Data_int.xlsx'
XDB=pd.read_excel(nxl,sheet_name=0)
XDB=XDB.dropna()
XDB.head(100)

#Se procede con la selección de las variables
XD=np.array(XDB[['College', 'FoodSpend', 'Income', 'TravelSpend']]) #Datos entrada
print(XD)
yd=np.array(XDB[['CreditCard']]) #Dato que quiero que el modelo pronostique
print(yd)

num_decimales = 5
# Use np.round to round NumPy arrays
XD_rounded = np.round(XD, num_decimales)
yd_rounded = np.round(yd, num_decimales)

print(XD_rounded)
print(yd_rounded)
```

2. Creación del modelo - Decision tree


```python
mar=DecisionTreeClassifier(criterion='gini', max_depth=4) #4 niveles o variables de entrada
mar.fit(XD,yd)
ydp=mar.predict(XD)

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(yd,ydp)
print(cm)

#Evaluamos el modelo
VN=cm[0,0]; FP=cm[0,1]; FN=cm[1,0]; VP=cm[1,1]

Ex=(VN+VP)/(VN+VP+FN+FP) #Que porcentaje de los datos pronostico correctamente
print("La exactitud alcanzada por el modelo es:", Ex)

Se=VP/(VP+FN) #Que porcentaje de los datos positivos pronosticados correctamente
print("La sensibilidad alcanzada por el modelo es:", Se)

Es=VN/(VN+FP) #Que porcentaje de los datos negativos pronosticados correctamente
print("La especificidad alcanzada por el modelo es:", Es)

#Este modelo tiene preferencia por los pre-negados
```

3. ¿Y, donde esta el arbol?

```python
#Librerias para graficar el árbol
from sklearn.tree import export_graphviz
from pydotplus import graph_from_dot_data

vs=['College', 'FoodSpend', 'Income', 'TravelSpend'] #Titulos para los nodos
dot_graph=export_graphviz(mar, feature_names=vs)
graph=graph_from_dot_data(dot_graph)
graph.write_png('arbol.png')
```

4. Se procede con el pronostico de los nuevos solicitantes

```python
XDB2=pd.read_excel(nxl,sheet_name=1)
XDB2=XDB2.dropna() #Elimina celdas vacias
XDB2.head(100)

#Asumiendo 'College' es la cloumna con 'Yes'/'No' values
XDB2['College'] = XDB2['College'].map({'Yes': 1, 'No': 0}) # Convert 'Yes'/'No' to 1/0
print(XDB2)

XD2=np.array(XDB2[['College', 'FoodSpend', 'Income', 'TravelSpend']]) #Datos entrada
print(XD2)

#Pronosticamos la aprobacion
ydp2=mar.predict(XD2)
print(ydp2)

print("La probabilidad de que el primer cliente tenga planes de viajar es de:", np.sum(ydp2))
print("El numero total de datos es:", len(ydp2))
print("La probabilidad de que el segundo cliente tenga planes de viajar es de:", len(ydp2)-np.sum(ydp2))
```
